```python
df = pd.read_csv('userbehaviour.csv')

# Check for null values
print(df.isnull().sum())

# Get column information
print(df.info())

# Get descriptive statistics
print(df.describe())
```

```python
# Highest, lowest, and average screen time
highest_screen_time = df['Average Screen Time'].max()
lowest_screen_time = df['Average Screen Time'].min()
average_screen_time = df['Average Screen Time'].mean()

print(f"Highest Screen Time: {highest_screen_time}")
print(f"Lowest Screen Time: {lowest_screen_time}")
print(f"Average Screen Time: {average_screen_time}")
```

```python
# Highest, lowest, and average amount spent
highest_amount_spent = df['Average Spent on App (INR)'].max()
lowest_amount_spent = df['Average Spent on App (INR)'].min()
average_amount_spent = df['Average Spent on App (INR)'].mean()

print(f"Highest Amount Spent: {highest_amount_spent}")
print(f"Lowest Amount Spent: {lowest_amount_spent}")
print(f"Average Amount Spent: {average_amount_spent}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Separate active and uninstalled users
active_users = df[df['Status'] == 'Installed']
uninstalled_users = df[df['Status'] == 'Uninstalled']

# Plot the relationship
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Average Screen Time', y='Average Spent on App (INR)', data=active_users, label='Active Users')
sns.scatterplot(x='Average Screen Time', y='Average Spent on App (INR)', data=uninstalled_users, label='Uninstalled Users', color='red')
plt.title('Relationship between Spending and Screen Time')
plt.xlabel('Average Screen Time')
plt.ylabel('Average Spent on App (INR)')
plt.legend()
plt.show()
```

```python
# Plot the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Average Screen Time', y='Ratings', data=df)
plt.title('Relationship between Ratings and Average Screen Time')
plt.xlabel('Average Screen Time')
plt.ylabel('Ratings')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Prepare the data for clustering
features = df[['Average Screen Time', 'Average Spent on App (INR)', 'Ratings']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4)  # You can experiment with different numbers of clusters
df['Cluster'] = kmeans.fit_predict(scaled_features)

# Number of segments
num_segments = len(df['Cluster'].unique())
print(f"Number of Segments: {num_segments}")
```

```python
# Visualize the segments
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Average Screen Time', y='Average Spent on App (INR)', hue='Cluster', data=df, palette='viridis')
plt.title('User Segmentation Using K-Means Clustering')
plt.xlabel('Average Screen Time')
plt.ylabel('Average Spent on App (INR)')
plt.legend(title='Cluster')
plt.show()
```
